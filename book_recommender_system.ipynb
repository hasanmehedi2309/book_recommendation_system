# Popularity based recommender system

```python
ratings_with_name = ratings_df.merge(books_df,on="ISBN")
ratings_with_name
```

```python
ratings_with_name["Image-URL-S"][0]
```

```python
num_rating_df = ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating_df.rename(columns = {"Book-Rating":"num_rating"},inplace= True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name[["Book-Title","Book-Rating"]].groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_rating_df.rename(columns = {"Book-Rating":"avg_rating"},inplace= True)
avg_rating_df["avg_rating"] =  round(avg_rating_df["avg_rating"],2)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on= "Book-Title")
popular_df
```

```python
num_rating_df["num_rating"]>250
```

```python
popular_df = popular_df[popular_df["num_rating"]>=250].sort_values("avg_rating",ascending = False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books_df,on = "Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M","num_rating","avg_rating"]]
popular_df
```

```python
a = popular_df[["Book-Title","Image-URL-M"]]
a.iloc[0]
```

# Colaborative filtering based recommender system

```python
x = ratings_with_name.groupby("User-ID").count()["Book-Rating"]>200
knowledgable_user = x[x].index
knowledgable_user
```

```python
filtered_rating = ratings_with_name[ratings_with_name["User-ID"].isin(knowledgable_user)]
filtered_rating
```

```python
y = filtered_rating.groupby("Book-Title").count()["Book-Rating"]>=50
y[y].index
final_ratings = filtered_rating[filtered_rating["Book-Title"].isin(y[y].index)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index = "Book-Title", columns = "User-ID", values = "Book-Rating")
pt.fillna(0,inplace = True)
pt
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), reverse = True,key = lambda x:x[1])[:6]
    data = []
    for item in similar_items:
        temp_df = books_df[books_df["Book-Title"]==pt.index[item[0]]]
        data.extend(temp_df.drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M"]].values.tolist())
    return data
    
```

```python
recommend("1984")
```
